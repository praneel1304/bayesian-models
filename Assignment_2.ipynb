## Part 1: Continuous random variables

#### **Question 1.1:**
Margins (in cm) = $\{5.76, 4.56, 2.33, 6.32, 5.4\}$

Sample space ($\Omega$):- Experiment consists of 5 trials;

$$\Omega = \{ (m_1, m_2, m_3, m_4, m_5) \mid m_i \in \mathbb{R}, m_i \ge 0 \}$$

Each $m_i$ represents the margin of miss in cm for $i^{th}$ trial.

(a) Success :- $m_i \le 5$

(a) Failure :- $m_i > 5$

Let $\omega = (m_1, \dots, m_5)$ be an element of $\Omega$.

Indicator Function ($I_i$):-

$$I(m_i) = \begin{cases} 1 & m_i \le 5 \\ 0 & m_i > 5 \end{cases}$$

Random variable fn :- $X(\omega) = \sum_{i=1}^{5} I(m_i)$
$X(\omega)$ counts number of successful trials out of 5.

Value of $X$ for our data:-
No. of Success :- $\{2.33, 4.56\} = 2$
Failure :- $\{5.76, 6.32, 5.4\} = 3$

$$\boxed{\text{Discrete Sample Space of } X : \Omega_X = \{0, 1, 2, 3, 4, 5\}}$$

(b) Probability distribution of $X$

Each shot is independent.
Prob. Success in 1 shot = $p$

$$P(X=k) = \binom{5}{k} p^k (1-p)^{5-k}, \quad k \in \Omega_X$$

Where $k$ is no. of success in 5 trials.

## Part 3: The likelihood function
Data ($x_{obs}$): $303.25, 443, 220, 560, 880$.
Model :- $f(x, \mu) = \frac{1}{x\sqrt{2\pi}} e^{-\frac{(\ln x - \mu)^2}{2}}$

#### Question 3.1 (a)
Likelihood ($\mathcal{L}$) for fixed $x=220$ :-

$$\mathcal{L}(\mu) = f(220, \mu) = \frac{1}{220\sqrt{2\pi}} e^{-\frac{(\ln 220 - \mu)^2}{2}}$$

Function peaks at $\mu = \ln 220 \approx 5.39$.

#### Plot 3.1(a):

```r
log_normal_pdf <- function(x, mu) {
  (1 / (x * sqrt(2 * pi))) * exp(-((log(x) - mu)^2) / 2)
}

mus <- seq(3, 7, length.out = 200)
likelihood <- sapply(mus, function(m) log_normal_pdf(220, m))

plot(mus, likelihood, type = "l",
     xlab = expression(mu),
     ylab = "Likelihood",
     main = "Likelihood function for x = 220")

abline(v = log(220), lty = 2)
```

#### **Question 3.1 (b)**
Likelihood for Sample ($x_{obs}$) :-
$$\mathcal{L}(\mu) = \prod_{i=1}^{5} f(x_i, \mu)$$

#### Plot 3.1(b):



```r
x_obs <- c(303.25, 443, 220, 560, 880)

# Plot for full sample
likelihood_total <- sapply(mus, function(m) prod(log_normal_pdf(x_obs, m)))
plot(mus, likelihood_total, type='l', col='green',
     main='Likelihood for Full Sample', xlab='mu', ylab='Likelihood')

# Mark maximum
best_mu <- mus[which.max(likelihood_total)]
abline(v=best_mu, col='red', lty=2)
```
